### Exercício 1
$$ \lim_{x \to 0} \frac{\sqrt{x+9} - 3}{x} $$
Multiplicar pelo conjugado:
$$ \mathrm\:\frac{\sqrt{x+9}+3}{\sqrt{x+9}+3} $$ = $$ \frac{\left(\sqrt{x+9}-3\right)\left(\sqrt{x+9}+3\right)}{x\left(\sqrt{x+9}+3\right)} $$

$$ =\frac{x}{x\left(\sqrt{x+9}+3\right)} $$

Elimina o fator x em comum:
$$=\frac{1}{\sqrt{x+9}+3}$$
Dando:

$$ = \lim _{x\to \:0}\left(\frac{1}{\sqrt{x+9}+3}\right)$$
Insere o valor 0:
$$=\frac{1}{\sqrt{0+9}+3}$$
Simplificando:
$$=\frac{1}{6}$$






```python
import matplotlib.pyplot as plt
import numpy as np

# Define os valores de x
x = [0, 50, 100, -50, -100]

# Define o valor de y como 1/6 para todos os pontos de x
y = np.full(len(x), 1/6)

# Cria o gráfico
plt.plot(x, y)

# Define os títulos do gráfico e dos eixos
plt.title('Gráfico')
plt.xlabel('x')
plt.ylabel('y')

# Mostra o gráfico
plt.show()
```

### Exercício 2

$$ \lim_{x \to \infty} \frac{25x + 7}{\sqrt{5x^2 + 2}} $$

Dividir pelo denominador de maior potência:

$$\sqrt{5x^2+2}=x\sqrt{5+\frac{2}{x^2}}$$ $$ =\frac{25x+7}{x\sqrt{5+\frac{2}{x^2}}}$$
Divide isso por x:
$$=\frac{\frac{25x}{x}+\frac{7}{x}}{\frac{x\sqrt{5+\frac{2}{x^2}}}{x}}$$
Simplifica: $$=\frac{25+\frac{7}{x}}{\sqrt{5+\frac{2}{x^2}}}$$

Assim fazendo os proximos passos:


$$\lim _{x\to \infty \:}\left(25+\frac{7}{x}\right) = 25$$

$$ \lim _{x\to \infty \:}\left(\sqrt{5+\frac{2}{x^2}}\right) = \sqrt {5}$$

$$ \frac{25}{\sqrt{5}} = 5\sqrt{5}$$  



```python
import matplotlib.pyplot as plt
import numpy as np

# Cria um array de valores para x
x = np.linspace(-10, 10, 400)

# Calcula os valores de y para cada x usando a função
y = (25*x + 7) / np.sqrt(5*x**2 + 2)

# Cria o gráfico
plt.plot(x, y)

# Define os títulos do gráfico e dos eixos
plt.title('Gráfico da função (25x + 7) / sqrt(5x^2 + 2)')
plt.xlabel('x')
plt.ylabel('y')

# Mostra o gráfico
plt.show()
```

##Exercício 3
$$
\lim_{x \to 1} f(x) \text{ onde } f(x) = \begin{cases}
2x+1 & x \leq 1 \\
x^2 + 1 & x > 1
\end{cases}
$$

Vamos calcular o limite de \( f(x) \) quando \( x \to 1 \).

A função \( f(x) \) é definida por partes:

\
f(x) =
\begin{cases}
2x + 1 & \text{se} \, x \leq 1 \\
x^2 + 1 & \text{se} \, x > 1
\end{cases}


Vamos calcular o limite pela esquerda de 1, usando a definição de \( f(x) = 2x + 1 \) para \( x \leq 1 \):

$$\lim_{x \to 1^-} f(x) = \lim_{x \to 1^-} (2x + 1) = 3.$$

Agora, vamos calcular o limite pela direita de 1, usando a definição de $$ f(x) = x^2 + 1\  para\  x > 1 $$

$$\lim_{x \to 1^+} f(x) = \lim_{x \to 1^+} (x^2 + 1) = 2.$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Cria um array de valores para x
x = np.linspace(0, 2, 400)

# Calcula os valores de y para cada x usando a função
y = np.piecewise(x, [x <= 1, x > 1], [lambda x: 2*x + 1, lambda x: x**2 + 1])

# Cria o gráfico
plt.plot(x, y)

# Adiciona pontos para os limites laterais
plt.plot(1, 3, 'ro') # Limite pela esquerda
plt.plot(1, 2, 'go') # Limite pela direita

# Define os títulos do gráfico e dos eixos
plt.title('Gráfico da função com limites laterais')
plt.xlabel('x')
plt.ylabel('y')

# Mostra o gráfico
plt.show()
```

##Exercício 5
$$ \lim_{x \to \infty} \frac{4x^2 + 3x - 5}{2x^2 - x + 1} $$

Dividir pelo denominador de maior potência:
$$\frac{\frac{4x^2}{x^2}+\frac{3x}{x^2}-\frac{5}{x^2}}{\frac{2x^2}{x^2}-\frac{x}{x^2}+\frac{1}{x^2}}$$

Simplifica:
$$=\frac{4+\frac{3}{x}-\frac{5}{x^2}}{2-\frac{1}{x}+\frac{1}{x^2}}$$

Excessão da forma indeterminada:
$$=\frac{\lim _{x\to \infty \:}\left(4+\frac{3}{x}-\frac{5}{x^2}\right)}{\lim _{x\to \infty \:}\left(2-\frac{1}{x}+\frac{1}{x^2}\right)}$$

Resolução das partes:
$$\lim _{x\to \infty \:}\left(4+\frac{3}{x}-\frac{5}{x^2}\right)$$
$$=\lim _{x\to \infty \:}\left(4\right)+\lim _{x\to \infty \:}\left(\frac{3}{x}\right)-\lim _{x\to \infty \:}\left(\frac{5}{x^2}\right)$$
Resolvendo os limites: $$\lim _{x\to \infty \:}\left(4\right) = 4$$

$$\lim _{x\to \infty \:}\left(\frac{3}{x}\right) = 0$$

$$\lim _{x\to \infty \:}\left(\frac{5}{x^2}\right) = 0$$

Resultado:
$$ 4 + 0 - 0 = 4 $$

Resolvendo as partes: $$\lim _{x\to \infty \:}\left(2-\frac{1}{x}+\frac{1}{x^2}\right)$$
$$=\lim _{x\to \infty \:}\left(2\right)-\lim _{x\to \infty \:}\left(\frac{1}{x}\right)+\lim _{x\to \infty \:}\left(\frac{1}{x^2}\right)$$
Resolução os limites:
$$\lim _{x\to \infty \:}\left(2\right) = 2$$
$$\lim _{x\to \infty \:}\left(\frac{1}{x}\right) = 0$$
$$\lim _{x\to \infty \:}\left(\frac{1}{x^2}\right)=0$$
Resultado:
$$2 - 0 + 0 = 2$$


Resultado final:
$$\frac{4}{2} =2$$


```python
import matplotlib.pyplot as plt
import numpy as np

# Cria um array de valores para x
x = np.linspace(-10, 10, 400)

# Calcula os valores de y para cada x usando a função
y = (4*x**2 + 3*x - 5) / (2*x**2 - x + 1)

# Cria o gráfico
plt.plot(x, y)

# Define os títulos do gráfico e dos eixos
plt.title('Gráfico da função (4x^2 + 3x - 5) / (2x^2 - x + 1)')
plt.xlabel('x')
plt.ylabel('y')

# Mostra o gráfico
plt.show()
```
